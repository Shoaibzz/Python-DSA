```python
##while loop

count=0
while count<5:
    print(count)
    count=count+1
```

```python
count=0
while count%2==0:
    print(count)
    count=count+1
```

```python
##loop control statements

##break
##the break statement exits the loop prematurely

##break statement

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
##continue
##the continue statement skips the current iteration aand continues to next

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
##pass
##the pass statement is a null operation: it does nothing.

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
##nested loops
#3a loop inside a loop
count=0
for i in range(3):
    for i in range(2):
        count=count+1
        print(f"shoaib",count)
```

```python
##example calculating the first n natural numbers

##while loop

n=10
sum=0
count=0
while count<=n:
    sum=sum+count
    count=count+1

print("the sum is ",sum)
```

```python
sum=0
for i in range(11):
    sum=sum+i

print("the sum",sum)
```

```python
##example  -- prime numbers between 1 to101

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
