Sets

sets are built -in data type they do not allow duplicate elements. sets are useful for membership tests,eliminating dupicate entries and performing mathematical set operations like union,intersection,difference, and symmetric difference

```python
##create a set
my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
set1={1,2,3,4,4,4,2,2,2,2,4,5,6,6}
print(set1)
```

```python
set1.remove(3)
set1
```

```python
set1.discard(3) ## does not give any error
set1
```

```python
##pop method
set1.pop()
print(set1) ## the set is unordered list so it can remove any number
```

```python
##set memebership test
set2={1,2,3,4,5}
print(3 in set2)
```

```python
##mathematicasl operation
set1={1,2,3,4,5,6}
set2={5,6,7,8,9,0}
union_set=set1.union(set2)
print(union_set)
```
